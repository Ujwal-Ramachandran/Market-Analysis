```python
files = [file for file in os.listdir('C:/datasets/Market_data')]

all_months_data = pd.DataFrame()

for file in files:
    df = pd.read_csv('C:/datasets/Market_data/' + file)
    all_months_data = pd.concat([all_months_data, df])
    
```
